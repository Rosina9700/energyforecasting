```python
import warnings
import itertools
import pandas as pd
import plotting as mp
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import data_wrangling as dw
from baseline_models import Baseline_previous, Baseline_average
plt.style.use('fivethirtyeight')
```

```python
project_name = 'project_6d8c'
filelocation='../../capstone_data/Azimuth/clean/{}_featurized.csv'.format(project_name)
df = pd.read_csv(filelocation)
df['t'] = pd.to_datetime(df['t'], format='%Y-%m-%d %H:%M:%S')
df.set_index('t',inplace=True)
```

```python
df['power_1'] = df['load_v1rms'] * df['load_i1rms'] 
df['power_2'] = df['load_v2rms'] * df['load_i2rms'] 
df['power_3'] = df['load_v3rms'] * df['laod_i3rms'] 
df['power_all'] = (df['power_1'] +df['power_2']+df['power_3'])*(5./12)/1000
y = pd.DataFrame(df['power_all'])
y = y['power_all'].resample('H').mean()
y = y.fillna(y.bfill())
y = pd.DataFrame(y)
```

```python
y_plot = pd.DataFrame(y['power_all'][-336:])
y_plot['power_all'].plot(figsize=(15, 6))
plt.title('Timeseries for 2 weeks at facility 6d8c')
plt.xlabel('Date')
plt.ylabel('Energy demand (kWh)')
plt.show()
```

## Hour ahead forecasting

```python
# Forecasting last two days
y_true = y_plot[-48:].values
```

### Baseline model - previous

```python
preds = np.empty(len(y_true))
for i in xrange(-48,0):
    preds[i+1] = y_plot['power_all'].values[i]
```

```python
plt.figure(figsize=(15, 6))
plt.plot(y_true, label='True')
plt.plot(preds, label='Predicted')
plt.title('Forecast and True demand for last 2 days')
plt.xlabel('Date')
plt.ylabel('Energy demand (kWh)')
plt.legend(loc='best')
plt.show()
```

### Baseline model - average

```python
ba = Baseline_average()
forecasts = []
for i in xrange(-48,0):
    y_train = y[:i]
    model = ba.fit(y_train)
    forecasts.append(model.forecast(steps=1))
forecasts = np.asarray(forecasts)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(y_true, label='True')
plt.plot(forecasts, label='Predicted')
plt.title('Forecast and True demand for last 2 days')
plt.xlabel('Date')
plt.ylabel('Energy demand (kWh)')
plt.legend(loc='best')
plt.show()
```

## Tuned Sarima

```python
params = (3,0,2)
seasonal_params = (1,0,1,24)

mod = sm.tsa.statespace.SARIMAX(y,order=params,seasonal_order=seasonal_params,
                                enforce_stationarity=False,enforce_invertibility=False)
```

```python
results = mod.fit()
pred = results.get_prediction(start=y[-48:].index[0], dynamic=False)
```

```python
f = pred.predicted_mean
```

```python
plt.figure(figsize=(15, 6))
plt.plot(y_true, label='True')
plt.plot(f.values, label='Predicted')
plt.title('Forecast and True demand for last 2 days')
plt.xlabel('Date')
plt.ylabel('Energy demand (kWh)')
plt.legend(loc='best')
plt.show()
```
