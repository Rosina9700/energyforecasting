```python
import warnings
import itertools
import pandas as pd
import plotting as mp
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from baseline_models import Baseline_previous, Baseline_average
plt.style.use('fivethirtyeight')
```

```python
project_name = 'project_5526'
filelocation='../../capstone_data/Azimuth/clean/{}_featurized.csv'.format(project_name)
df = pd.read_csv(filelocation)
df['t'] = pd.to_datetime(df['t'], format='%Y-%m-%d %H:%M:%S')
df.set_index('t',inplace=True)
```

```python
df['power_1'] = df['load_v1rms'] * df['load_i1rms'] 
df['power_2'] = df['load_v2rms'] * df['load_i2rms'] 
df['power_3'] = df['load_v3rms'] * df['laod_i3rms'] 
df['power_all'] = (df['power_1'] +df['power_2']+df['power_3'])*(5./12)/1000
y = pd.DataFrame(df['power_all'])
y = y.fillna(y.bfill())
y = y['power_all'].resample('H').sum()
y = y.fillna(y.bfill())
y = pd.DataFrame(y)
```

```python
y['year'] = y.index.year
mp.plot_all_data(y,['power_all'])
```

```python
length_data = len(y)
y_train = y.ix[:-48,'power_all']
y_test = y.ix[-48:,'power_all']
```

```python
params = (1,0,2)
seasonal_params = (0,0,0,24)
mod = sm.tsa.statespace.SARIMAX(y_train,order=params,seasonal_order=seasonal_params,
                                enforce_stationarity=False,enforce_invertibility=False)
results = mod.fit()
print('ARIMA{}x{} - AIC:{}'.format(params, seasonal_params, results.aic))
```

```python
results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
pred = results.get_prediction(start=y_train.index.min(), dynamic=False)
pred_ci = pred.conf_int()
```

```python
y['predictions'] = pred.predicted_mean
```

```python
mp.plot_all_data(y, ['power_all','predictions'])
```

```python
ax = y['power_all'].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.8, figsize=(12,5))

# ax.fill_between(pred_ci.index,
#                 pred_ci.iloc[:, 0],
#                 pred_ci.iloc[:, 1], color='k', alpha=.2)

ax.set_xlabel('Date')
ax.set_ylabel('Demand')
plt.legend()

plt.show()
```

```python
pred_dynamic = results.get_prediction(start=y_train[-24:].index[0], dynamic=True, full_results=True)
pred_dynamic_ci = pred_dynamic.conf_int()
```

```python
ax = y_train[-168:].plot(label='observed', figsize=(20, 15))
pred_dynamic.predicted_mean.plot(label='Dynamic Forecast', ax=ax)

ax.fill_between(pred_dynamic_ci.index,
                pred_dynamic_ci.iloc[:, 0],
                pred_dynamic_ci.iloc[:, 1], color='k', alpha=.25)

ax.fill_betweenx(ax.get_ylim(), pd.to_datetime('2017-06-15'), y.index[-1],
                 alpha=.1, zorder=-1)

ax.set_xlabel('Date')
ax.set_ylabel('Demand')

plt.legend()
plt.show()
```

```python
def rolling_predictions_sarima(y,end,window,params):
    forecast = np.zeros(window)
    for i in xrange(window):
        y_temp = y[0:end+i] 
        model = sm.tsa.statespace.SARIMAX(y_temp,order=params[0],seasonal_order=params[1],
                                enforce_stationarity=False,enforce_invertibility=False)
        model = model.fit()
        forecast[i]= model.forecast(steps=1).values[0]
    return forecast
```

```python
y.drop('predictions',axis=1,inplace=True)
```

```python
params = (1,0,2)
seasonal_params = (0,0,0,24)
parameters = (params, seasonal_params)
forecast = rolling_predictions_sarima(y['power_all'], len(y_train), 48, parameters)
```

```python
plt.figure(figsize=(15,8))
plt.plot(y_test.values, label='measured')
plt.plot(forecast,label='hour ahead forecast')
plt.legend(loc='best')
plt.show()
```
