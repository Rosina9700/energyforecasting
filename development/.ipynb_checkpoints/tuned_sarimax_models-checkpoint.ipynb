```python
import warnings
import itertools
import pandas as pd
import plotting as mp
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from baseline_models import Baseline_previous, Baseline_average
plt.style.use('fivethirtyeight')
```

```python
project_name = 'project_1074'
filelocation='../../capstone_data/Azimuth/clean/{}_featurized.csv'.format(project_name)
df = pd.read_csv(filelocation)
df['t'] = pd.to_datetime(df['t'], format='%Y-%m-%d %H:%M:%S')
df.set_index('t',inplace=True)
```

```python
df['power_1'] = df['load_v1rms'] * df['load_i1rms']
df['power_2'] = df['load_v2rms'] * df['load_i2rms']
df['power_3'] = df['load_v3rms'] * df['laod_i3rms']
df['power_all'] = df['power_1'] +df['power_2']+df['power_3'] * 5./12
```

```python
mp.plot_all_data(df,['power_all'])
```

```python
y = df['power_all']
y = y.fillna(y.bfill())
y = pd.DataFrame(y.resample('H').sum())
exog = df['T']
exog = exog.fillna(exog.bfill())
exog = pd.DataFrame(exog.resample('H').mean())
y['T-1'] = exog['T'].shift(1)
y = y.fillna(y.bfill())
y['weekday'] = y.index.dayofweek
y['weekday'] = y['weekday'].apply(lambda x: 1 if x < 5 else 0)
y_train = y[:-24]
y_test = y[-24:]
```

```python
endog = y_train.ix[:,0]
```

```python
exog = sm.add_constant(y_train.ix[:,1:])
```

```python
params = ((0,1,1),(1,0,1,24))
model = sm.tsa.statespace.SARIMAX(endog, exog,
                                    order=params[0],
                                    seasonal_order=params[1],
                                    enforce_stationarity=False,
                                    enforce_invertibility=False)
model=model.fit()
```

```python
model.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
pred = model.get_prediction(start=y_train.index.min(), dynamic=False)
pred_ci = pred.conf_int()
```

```python
y['predictions'] = pred.predicted_mean
```

```python
y['year'] = y.index.year
mp.plot_all_data(y[1:], ['power_all','predictions'])
```

```python
pred_dynamic = model.get_prediction(start=y_train[-24:].index[0], dynamic=True, full_results=True)
pred_dynamic_ci = pred_dynamic.conf_int()
```

```python
ax = y_train[-168:]['power_all'].plot(label='observed', figsize=(20, 15))
pred_dynamic.predicted_mean.plot(label='Dynamic Forecast', ax=ax)

ax.fill_between(pred_dynamic_ci.index,
                pred_dynamic_ci.iloc[:, 0],
                pred_dynamic_ci.iloc[:, 1], color='k', alpha=.25)

ax.fill_betweenx(ax.get_ylim(), pd.to_datetime('2017-06-15'), y.index[-1],
                 alpha=.1, zorder=-1)

ax.set_xlabel('Date')
ax.set_ylabel('Demand')

plt.legend()
plt.show()
```

```python
def rolling_predictions_sarima(y,end,window,params):
    forecast = np.zeros(window)
    endog = y.ix[:,0]
    exog = sm.add_constant(y.ix[:,1:])
    for i in xrange(window):
        endog_temp = endog.ix[:end+i]
        exog_temp =  exog.ix[:end+i,:]
        try:
            model = sm.tsa.statespace.SARIMAX(endog_temp, exog_temp,
                                    order=params[0],
                                    seasonal_order=params[1],
                                    enforce_stationarity=False,
                                    enforce_invertibility=False)
            model=model.fit()
        except:
            print 'SKIPPED {}-{}'.format(params[0], params[1])
            model = None
            break
        forecast[i]= model.forecast(steps=1,exog=exog.ix[end+i,:].values.reshape(1,exog.shape[1])).values[0]
    true = endog[end:end+window].values
    rmse = np.sqrt(((true-forecast)**2).mean())
    return forecast
```

```python
y.drop(['year'], axis=1, inplace=True)
```

```python
y.drop(['predictions'], axis=1, inplace=True)
y.head()
```

```python
params = ((0,1,1),(1,0,1,24))
forecast = rolling_predictions_sarima(y, len(y_train), 24, params)
```

```python
plt.figure(figsize=(15,8))
plt.plot(y_test[:24]['power_all'].values, label='measured')
plt.plot(forecast,label='hour ahead forecast')
plt.legend(loc='best')
plt.show()
```
